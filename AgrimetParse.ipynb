```python
def generate_url(start_year, start_month, start_day, end_year, end_month, end_day):
    # Station and pcodes (parameter codes)
    station = "FOGO"
    pcodes = ["OBM", "TU", "WS", "SI", "PC"]

    # Base URL
    base_url = "https://www.usbr.gov/pn-bin/instant.pl?"

    # Construct the URL with parameters
    url = f"{base_url}station={station}&year={start_year}&month={start_month}&day={start_day}&year={end_year}&month={end_month}&day={end_day}"

    # Add the parameter codes (pcodes) to the URL
    for pcode in pcodes:
        url += f"&pcode={pcode}"

    # Add the 'print_hourly' parameter
    url += "&print_hourly=1"

    return url
```

```python
# Example usage:
end_year = 2023
end_month = 12
end_day = 26

# Generate the URL
url = generate_url(start_year, start_month, start_day, end_year, end_month, end_day)

print(url)
```

```python
# GET request to get the URL
response = requests.get(url)
```

```python
# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content
    html_content = response.text

    # Create a BeautifulSoup object
    soup = BeautifulSoup(html_content, 'html.parser')

    # Find all <pre> elements 
    pre_elements = soup.find_all('pre')

    # Check if there are two <pre> elements
    if len(pre_elements) >= 2:
        # Get the content of the second <pre> element
        pre_element = pre_elements[1]

        # Split the text by lines
        lines = pre_element.text.strip().split('\n')

        # Split the lines into columns based on comma 
        data = [line.split(',') for line in lines]
    else:
        print("Unable to find the 2nd <pre> element.")
else:
    print(f"Failed to retrieve data from {url}. Status code: {response.status_code}")
```

```python
# Create a pandas DataFrame
new_df = pd.DataFrame(data)[3:-1]

# Rename the columns
new_df.columns = ["DATETIME", "FOGO OBM", "FOGO TU", "FOGO WS", "FOGO SI", "FOGO PC"]

new_df
```
