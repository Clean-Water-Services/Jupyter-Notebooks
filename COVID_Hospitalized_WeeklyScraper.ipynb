## Extract Data

###  Data by Facility

### Extract data from tableu dashaboard

```python
#Create Tabluea Scrapper object and load url into it
ts_zip = TS()
ts_zip.loads(ByFacility)

#Get worksheet data
workbook = ts_zip.getWorkbook()
for t in workbook.worksheets:
    print(f"worksheet name : {t.name}") #show worksheet name
    print(t.data) #show dataframe for this worksheet
```

```python
# Dictionary to store data from all worksheets
dataframes = {}

# Loop through all worksheets in the workbook
for worksheet in workbook.worksheets:
    # Store the data in the dictionary with the worksheet name as the key
    dataframes[worksheet.name] = worksheet.data
    print(worksheet.name)

# Now, dataframes contains a DataFrame for each worksheet
# For example, to access data from the 'Patients by Severity' worksheet:
#dataframes['Patients by Severity']
```

```python
df = dataframes['7 Day Average of COVID-19 Positive Patients by Facility']
df.head(3)
```

```python
for col in df.columns:
    print(col)
    print(df[col].unique())
```

### Prepare new dataframe

```python
df_covid_cases=pd.DataFrame()
df_covid_cases['7 Day Average of COVID-19 Positive Patients'] = df['7 Day Average of COVID-19 Positive Patients-alias']
df_covid_cases['MAX_7 Day Average of COVID-19 Positive Patients_Num'] = df_covid_cases['7 Day Average of COVID-19 Positive Patients'].replace('<10', '9').astype(int)
df_covid_cases['MIN_7 Day Average of COVID-19 Positive Patients_Num'] = df_covid_cases['7 Day Average of COVID-19 Positive Patients'].replace('<10', '1').astype(int)
df_covid_cases['County']= df['County-value']
df_covid_cases['Hospital Name']=df['Hospital Name-value']
```

```python
df_covid_cases
```

### Agreggate Values

```python
# Group by 'County' and sum both the max and min columns
aggregated_data = df_covid_cases.groupby('County').agg({
    'MAX_7 Day Average of COVID-19 Positive Patients_Num': 'sum',
    'MIN_7 Day Average of COVID-19 Positive Patients_Num': 'sum'
}).reset_index()
```

```python
# Given string
s = dataframes['UpdatedTimestamp']['Column Labels-alias'][0]
# Use regular expression to extract the date
match = re.search(r'(\d{1,2}/\d{1,2}/\d{4})', s)
if match:
    date = match.group(1)
    print(date)  # Outputs: 8/8/2023
else:
    print("Date not found")
```

```python
aggregated_data['DateStamp']=date
aggregated_data.head(5)
```

```python
# Pivot the DataFrame
df_pivot = aggregated_data.pivot(index='DateStamp', columns='County', values=['MAX_7 Day Average of COVID-19 Positive Patients_Num', 'MIN_7 Day Average of COVID-19 Positive Patients_Num'])

# Adjust column levels
df_pivot.columns = df_pivot.columns.swaplevel(0, 1)
df_pivot.sort_index(axis=1, level=0, inplace=True)
```
